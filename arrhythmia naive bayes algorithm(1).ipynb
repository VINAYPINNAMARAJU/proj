```python
y = data.diagnosis.values
x = data.drop(["diagnosis"],axis=1)
```

Diagnosis values were loaded on y axis. The x axis was loaded with all data except diagnosis. The diagnosis feature is a feature of classes 1-16. 



The train_test_split function randomly divides the dataset into train and test. In other words, 60% train section is not 60% from the beginning of dataset. It is the amount corresponding to 60% of all data. The expression random_state = 42 allows the same 60% and 20% sections to be used each time this function is executed. In this way, the work produced by different people at different times produces the same output. 42 is the meaning of life in the book A Hitchhiker's Guide to the Galaxy. Instead of 42, 20, 19 can also be used.



```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2,random_state=25)
x_train, x_validate, y_train, y_validate = train_test_split(x_train,y_train,test_size=0.25, random_state=25)
```

By calling the Naive Bayes Algorithm from the SkitLearn library, the algorithm is trained with x_train and y_train values.



```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(x_train,y_train)
```

```python
x_test=data.head(2)
x_test=x_test.drop(["diagnosis"],axis=1)
```

```python
x_test
```

```python
y_pred = nb.predict(x_test)
print("Estimated Value: ",y_pred)
```

```python
from sklearn import metrics
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score
from sklearn.metrics import precision_score
print("Confussion Matrix: ",confusion_matrix(y_test, y_pred))
print("Accuracy rate: ",metrics.accuracy_score(y_test,y_pred))
print("f1 Score: ",f1_score(y_test, y_pred, average='macro'))
print("Accuracy Value: ",precision_score(y_test, y_pred, average='macro'))
```
